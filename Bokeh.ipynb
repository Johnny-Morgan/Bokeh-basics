# Task
Create a bar chart from the autompg2 dataset which shows the average city MPG 
for the models of cars found in rows 100-140.

# Annotations

```python
from bokeh.plotting import figure, output_notebook, show
from bokeh.models import ColumnDataSource, LabelSet

output_notebook()
data = {'height':[199, 176, 190, 140, 165, 182], 
        'weight':[75, 74, 80, 65, 68, 94], 
        'name':['Jacob', 'John', 'Jack', 'Kate', 'Sawyer', "Hurley"]}

source = ColumnDataSource(data)

p = figure(plot_width = 300, plot_height = 300, x_axis_label = 'Height (cm)', 
           y_axis_label = 'Weight (kg)', title = 'Heigth vs Weight', x_range= (130, 220), y_range= (60, 100))

p.circle(x='height', y='weight', source = source)

labels = LabelSet(x="height", y="weight", text= "name", source= source, x_offset = 5, y_offset = 5)
p.add_layout(labels)


show(p)
```

# Task
Amend your engine displacement vs city MPG plot to show car model names as labels. 
Do this for records 100-110 in autompg2


```python
from bokeh.plotting import figure, output_notebook, show
from bokeh.models import ColumnDataSource, LabelSet
from bokeh.sampledata.autompg2 import autompg2 as data
output_notebook()

source = ColumnDataSource(data[100:110])

p = figure(x_range=(1,3), y_range = (15,28), x_axis_label="Displacement", 
           y_axis_label = "mpg", title = "Displacement vs mpg")

p.circle(x= "displ", y= "cty", source = source)

labels = LabelSet(x= "displ", y= "cty", source = source, text = "model", x_offset = 5, y_offset = 5, text_font_size = "15pt")
p.add_layout(labels)

show(p)
```

# Linking & Brushing

```python
from bokeh.plotting import figure, output_notebook, show
from bokeh.layouts import gridplot

x = [1, 2, 3, 4]
y = [1, 2, 3, 4]
y2 = [4, 3, 2, 1]

p = figure(width = 200, height = 200, x_range = (0, 5), y_range = (0, 5), title = "Plot 1")

p2 = figure(width = 200, height = 200, x_range = p.x_range, y_range = p.y_range, title = "Plot 2")

p.circle(x, y)
p2.square(x, y2)

gplot = gridplot([[p, p2]])
show(gplot)
```

```python
from bokeh.plotting import figure, output_notebook, show
from bokeh.models import ColumnDataSource
from bokeh.sampledata.autompg2 import autompg2 as data
from bokeh.layouts import gridplot

x = [1, 2, 3, 4]
y = [1, 2, 3, 4]
x2 = [5, 3, 4, 2]
y2 = [4, 3, 2, 1]

source = ColumnDataSource(data=dict(x=x, y=y, x2=x2, y2=y2))

tools = "box_select", "pan"

p = figure(width = 200, height = 200, x_range = (0, 5), y_range = (0, 5), title = "Plot 1", tools = tools)

p2 = figure(width = 200, height = 200, x_range = p.x_range, y_range = p.y_range, title = "Plot 2", tools = tools)

p.circle("x", "y", source = source)
p2.square("x2", "y2", source = source)

gplot = gridplot([[p, p2]])
show(gplot)
```

```python
from bokeh.plotting import figure, output_notebook, show
from bokeh.models import ColumnDataSource, LabelSet
from bokeh.sampledata.autompg2 import autompg2 as data
output_notebook()

source = ColumnDataSource(data[100:110])
tools = "box_select", "pan"

p = figure(width = 500, height = 300, x_range=(1,3), y_range = (15,40), x_axis_label="Displacement", 
           y_axis_label = " highway mpg", title = "Displacement vs highway mpg", tools=tools)
p2 = figure(width = 500, height = 300, x_range=p.x_range, y_range = p.y_range, x_axis_label="Displacement", 
           y_axis_label = "city mpg", title = "Displacement vs city mpg", tools=tools)

p.circle(x= "displ", y= "hwy", source = source)
p2.circle(x= "displ", y= "cty", source = source)

labels = LabelSet(x= "displ", y= "hwy", source = source, text = "model", x_offset = 5, y_offset = 5, text_font_size = "10pt")
labels2 = LabelSet(x= "displ", y= "cty", source = source, text = "model", x_offset = 5, y_offset = 5, text_font_size = "10pt")
p.add_layout(labels)
p2.add_layout(labels2)

gplot = gridplot([p], [p2])
show(gplot)
```

```python
from bokeh.plotting import figure, output_notebook, show
from bokeh.models import ColumnDataSource, LabelSet
from bokeh.sampledata.autompg2 import autompg2 as data
output_notebook()

group = data.groupby("displ")


data_source = ColumnDataSource(data)
group_source = ColumnDataSource(group)
tools = "box_select", "pan"

p = figure(width = 500, height = 300, x_range=(1,8), y_range = (5,45), x_axis_label="Displacement", 
           y_axis_label = "highway mpg", title = "Displacement vs highway mpg", tools=tools)
p2 = figure(width = 500, height = 300, x_range=p.x_range, y_range = p.y_range, x_axis_label="Displacement", 
           y_axis_label = "city mpg", title = "Displacement vs city mpg", tools=tools)

p.circle(x= "displ", y= "hwy", source = data_source)
p2.circle(x= "displ", y= "cty", source = data_source)
p.line(x= "displ", y = "hwy_mean", source = group_source, color = "red")
p2.line(x= "displ", y = "cty_mean", source = group_source, color = "red")


gplot = gridplot([p], [p2])
show(gplot)
```
